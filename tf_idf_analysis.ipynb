```python
import tf_idf_observer as tid
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd

```

Test with more neutral data (inclusive)

```python
cm_labels = ["Control", "Low", "Med", "High"]
tfidf_all_neutral = []
tfidf_noctrl_neutral = []
tfidf_noctrl_middle_neutral = []

turn_index = []


for i in range(5):
    turn_index.append(i)
    tfidf_all_neutral.append(tid.tf_idf_accuracy('datasets/synthetic_conversations_full.csv', turn_index, False, False))
    tfidf_noctrl_neutral.append(tid.tf_idf_accuracy('datasets/synthetic_conversations_full.csv', turn_index, True, False))
    tfidf_noctrl_middle_neutral.append(tid.tf_idf_accuracy('datasets/synthetic_conversations_full.csv', turn_index, True, True))


#print(tfidf_all[0][0]*100, tfidf_all[1][0]*100, tfidf_all[2][0]*100, tfidf_all[3][0]*100, tfidf_all[4][0]*100)

tests = ['1', '2', '3', '4', '5']
scores = [tfidf_all_neutral[0][0]*100, tfidf_all_neutral[1][0]*100, tfidf_all_neutral[2][0]*100, tfidf_all_neutral[3][0]*100, tfidf_all_neutral[4][0]*100]
plt.bar(tests, scores)
plt.title("tf-idf with all neutral data (inclusive)")
plt.xlabel("Conversation Turns")
plt.ylabel("Accuracy %")
plt.ylim(0, 100)
plt.show()

for i in range(5):
    sb.heatmap(tfidf_all_neutral[i][1], annot=True, cmap='viridis', fmt='d', xticklabels=cm_labels, yticklabels=cm_labels)
    plt.title("Confusion Matrix of turn size " + str(i+1))
    plt.show()

for i in range(5):
    print(tfidf_all_neutral[i][0]*100)

cm_labels = ["Low", "Med", "High"]

tests = ['1', '2', '3', '4', '5']
scores = [tfidf_noctrl_neutral[0][0]*100, tfidf_noctrl_neutral[1][0]*100, tfidf_noctrl_neutral[2][0]*100, tfidf_noctrl_neutral[3][0]*100, tfidf_noctrl_neutral[4][0]*100]
plt.bar(tests, scores)
plt.title("tf-idf with no control neutral data (inclusive)")
plt.xlabel("Conversation Turns")
plt.ylabel("Accuracy %")
plt.ylim(0, 100)
plt.show()

for i in range(5):
    sb.heatmap(tfidf_noctrl_neutral[i][1], annot=True, cmap='viridis', fmt='d', xticklabels=cm_labels, yticklabels=cm_labels)
    plt.title("Confusion Matrix of turn size " + str(i+1))
    plt.show()

for i in range(5):
    print(tfidf_noctrl_neutral[i][0]*100)

cm_labels = ["Low", "High"]

tests = ['1', '2', '3', '4', '5']
scores = [tfidf_noctrl_middle_neutral[0][0]*100, tfidf_noctrl_middle_neutral[1][0]*100, tfidf_noctrl_middle_neutral[2][0]*100, tfidf_noctrl_middle_neutral[3][0]*100, tfidf_noctrl_middle_neutral[4][0]*100]
plt.bar(tests, scores)
plt.title("tf-idf with extreme neutral data (inclusive)")
plt.xlabel("Conversation Turns")
plt.ylabel("Accuracy %")
plt.ylim(0, 100)
plt.show()

for i in range(5):
    sb.heatmap(tfidf_noctrl_middle_neutral[i][1], annot=True, cmap='viridis', fmt='d', xticklabels=cm_labels, yticklabels=cm_labels)
    plt.title("Confusion Matrix of turn size " + str(i+1))
    plt.show()

for i in range(5):
    print(tfidf_noctrl_middle_neutral[i][0]*100)
```
